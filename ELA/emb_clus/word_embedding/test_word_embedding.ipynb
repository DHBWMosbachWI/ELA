```python
from multiprocessing import Pool
import numpy as np
import pandas as pd
import ast

def parallelize_dataframe(df, func, n_cores=4):
    df_split = np.array_split(df, n_cores)
    pool = Pool(n_cores)
    df = pd.concat(pool.map(func, df_split))
    pool.close()
    pool.join()
    return df

def apply_literal_eval(df):
    df["word_embedding"] = df["word_embedding"].apply(ast.literal_eval)
    return df
```

```python
def getAllStoredPickleFiles():
    pickleFilesToRead = []
    for (dirpath, dirnames, filenames) in os.walk("./"):
        if dirpath == "./out":
            for filename in filenames:
                if "storeEmbeddingResults" in filename:
                    pickleFilesToRead.append(filename)
    return pickleFilesToRead

def loadAllPickleFilesAndCombineToOne(pickleFilesToRead, tablesToRead=[]):
    storeEmbeddingResults = {"results": []}
    for pickleFile in pickleFilesToRead:
        currentResults = pickle.load(open("./out/"+pickleFile, "rb"))
        for result in currentResults["results"]:
            if len(tablesToRead) == 0:
                storeEmbeddingResults["results"].append(result)
            elif result['tablename'] in tablesToRead:
                storeEmbeddingResults['results'].append(result)
                
    print("Number of Tables: ", len(storeEmbeddingResults["results"]))
    return storeEmbeddingResults
```

```python
all_word_embeddings = loadAllPickleFilesAndCombineToOne(getAllStoredPickleFiles())
```

```python
all_word_embeddings_new = {}

for index, current_result in enumerate(all_word_embeddings["results"]):
    #print(current_result)
    tablename= current_result['tablename']

    all_word_embeddings_new[tablename] = {}

    with open(f"./out/word_embeddings_{tablename}.p", "wb") as file:
        pickle.dump(current_result, file)
    
    for key in current_result.keys():
        if "col" in key:
            all_word_embeddings_new[tablename][key] = current_result[key]

pickle.dump(all_word_embeddings_new, open(f"./out/word_embeddings_all.p", "wb"))
```
